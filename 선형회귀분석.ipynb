```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init

print(torch.__version__)
```

```python
num_data = 1000 # 데이터
num_epoch = 500 # 반복회수

# torch.nn.init.uniform_(출력할 텐서, 반복 시작, 반복 끝)
x = init.uniform_(torch.Tensor(num_data, 1), -10, 10) # uniform_ 함수는 파라미터을 통해 -10 ~ 10까지 균등, uniform하게 초기화 한다.

#선형방정식
y = 2*x + 3

# torch.nn.init.normal_(출력할 텐서, 평균(mean), 표준편차(std))
## normal의 뜻은 normal Distribution, 정규 분포로서, 초기화시 넣어준 파라미터들 평균과 표준편차를 만족하는 정규 분포 값을 계산하여 tensor에 넣는다ㅓ.
noise = init.normal_(torch.FloatTensor(num_data, 1), std=1) # y_noise와 같은 noise 텐서를 만든다, 

# noise를 추가해주는 이유는 실제 데이터는 noise가 포함돼있기 때문에 사전에 추가해준다ㅓ.
y_noise = 2*(x+noise) + 3
```

```python

# torch.nn.Linear() 함수를 사용하여 선형회귀 모델을 만든다.
## Linear(가중치(weight), 편차(bias))
model = nn.Linear(1, 1)

#선형회귀의 결과값과 y_noise의 차이를 구한다.
##L1Loss는 평균 절대 오차(MAE)를 구한다.
loss_func = nn.L1Loss()
```

```python

#torch.optim는 옵티마이저를 사용할 수 있도록 해준다.
## 옵티마이저는 최적화 함수라고 하며, 이번에 사용하는 SGD는 경사하강법이다.
### SGD, Stochatic Gradient Descent 한 번에 들어오는 데이터의 수대로 경사하강법 알고리즘을 적용하는 최적화 함수이다.
#### lr, learning rate로서 학습률이라고 한다.
optimizer = optim.SGD(model.parameters(), lr=0.01)  # 선형회귀 모델과 학습률 0.01을 부여한다.
```

```python
label = y_noise #노이즈가 추가된 y값을 가져온다

for i in range(num_epoch): #epoch만큼 반복한다.
  
  #현재 optimizer에 부여돼있는 기울기 w 값을 초기화 시킨다.
  ## 0으로 초기화 하여 새로운 개산을 하기 위해
  optimizer.zero_grad() 

  # 선형회귀 모델에 유니폼함수로 초기화한 데이터( -10 ~ 10 )를 넣어준다.
  output = model(x)

  # 선형회귀 결과값과 노이즈가 추가된 라벨 값을 L1loss function에 넣는다.
  loss = loss_func(output, label)

  #w, b에 대해 기울기가 계산된다.
  loss.backward()
  
  #옵티마이저에 넣어준 model.parameters()에서 리턴되는 변수들의 기울기에 학습률 0.01을 곱하여 뺴줘 업데이트 시킨다.
  optimizer.step()

  if i % 10 == 0:
    print(loss.data)

  aram_list = list(model.parameters())

  print(param_list[0].item() , param_list[1].item())
```
